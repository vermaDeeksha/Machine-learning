```python
# since the is a regression problem statement we will not check imbalanced dataset.
# feature Scaling
# Step1 segregate dependent and independent variables.
x=data.iloc[:,0:3] #independent variables
y=data.iloc[:,-1]  # dependent variable
```

```python
#  step 2 train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# applying standard scalar( normalization /Standardization)
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
```

```python
x_train=Scaler.fit_transform(x_train)
x_test=Scaler.transform(x_test)

```

```python
#  Model building
from sklearn.linear_model import LinearRegression
regression =LinearRegression()
```

```python
# training the model
regression.fit(x_train,y_train)
```

```python
# y=mx+c
# c is intercept
print(regression.intercept_)
```

```python
# finding coefficients
print(regression.coef_)
```

```python
# testing data
reg_pred=regression.predict(x_test)
```

```python
# checking accuracy
#  R-square value
from sklearn.metrics import r2_score
accuracy_score=r2_score(y_test,reg_pred)
```
