# Q5- Execute KMeans

```python
x=data1.iloc[:,2:].values # selecting the fetures

```

```python
x.shape
```

As there is a difference in magnitude of credit limit and other features, we will scale the features

As there are not any negative values, we can use MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaled_features=scaler.fit_transform(x)
```

```python
scaled_features
```

# Modeling

```python
from sklearn.cluster import KMeans
```

```python
#  writing a for loop to a plot the graph for the with in cluster sum of square

wcss=[] # with in cluster sum of square
for i in range(1,11):
    kmeans=KMeans(n_clusters=i ,init='k-means++',random_state= 42) # initailizing
    kmeans.fit(scaled_features) # fitting on scaled feature
    wcss.append(kmeans.inertia_)# appending the wscss to the bank list
plt.plot(range(1,11),wcss)    
plt.title("The Elbow method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# As there is smooth descent after n_cluster=3 we will choose number of cluster as 3
```

```python
kmeans2=KMeans(n_clusters=3, init='k-means++', random_state=0)
```

```python
y_kmeans=kmeans2.fit_predict(x)
```

# Q7- Calculate avg silhoutte scores

```python
from sklearn.metrics import silhouette_score
```

```python
kmeans_score=silhouette_score(scaled_features,y_kmeans)
kmeans_score
```

```python
hc_score=silhouette_score(scaled_features,y_hc)
hc_score
```

```python
print("The silhoutte score of kmeans and hierarchiacal cluster are {} and {} respectively".format(kmeans_score,hc_score))
```

The HC_score is greater than kmeans_score

It analysis is as follows-

+1 score- near +1 silhouette score indicates that the simple is far away from its neighboring cluster

0 score- 0 silhouette score indicates that tha sample is on or very close to the decision boundary separating two neighouring cluster


-1= 1 silhouette score indicates that the sample have been assigment to the wrong cluster
